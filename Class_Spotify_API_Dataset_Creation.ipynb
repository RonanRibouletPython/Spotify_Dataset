```python
class Spotify_API_Handler():
  # Spotify API connection
  def __init__(self, client_id, client_secret):
    self.client_id = client_id
    self.client_secret = client_secret
    self.client_credentials_manager = SpotifyClientCredentials(client_id=self.client_id, client_secret=self.client_secret)
    self.sp = spotipy.Spotify(client_credentials_manager=self.client_credentials_manager)

  # Get the artist URI
  def get_artist(self, link)->str:
    self.link = link

    return self.link.split("/")[-1].split("?")[0]

  # Get the artist name
  def get_artist_name(self, URI)-> str:
    self.name = self.sp.artist(URI)

    return self.name["name"]

  def get_artist_albums(self, URI)->dict:

    # Creation of lists to store data
    artist_albums_ids = []
    artist_albums_titles = []
    artist_albums_release_date = []
    artist_albums_cover_url = []

    self.URI = URI

    albums_infos = self.sp.artist_albums(artist_id=URI)

    for i in range(len(albums_infos["items"])):
      artist_albums_titles.append(albums_infos['items'][i]['name'])
      artist_albums_ids.append(albums_infos['items'][i]['uri'])
      artist_albums_release_date.append(albums_infos['items'][i]['release_date'])
      artist_albums_cover_url.append(albums_infos['items'][i]['images'][0]["url"])

    return {
            "title": artist_albums_titles,
            "id": artist_albums_ids,
            "release_date": artist_albums_release_date,
            "cover_url": artist_albums_cover_url
    }

  def get_tracks(self, albums)->dict:

    tracks_raw_infos = []

    # Display and store the track infos
    track_duration_ms = []
    track_explicit = []
    track_url = []
    track_id = []
    track_name = []
    track_type = []
    track_preview_url = []
    track_album = []
    track_number = []

    num_of_albums = len(albums["id"])

    # Store the track infos of all the albums
    for i in range(num_of_albums):
      store_temp = self.sp.album_tracks(album_id=albums["id"][i])

      tracks_raw_infos.append(store_temp)

    for i in range(num_of_albums):
      num_of_tracks = len(tracks_raw_infos[i]['items'])
      for y in range(num_of_tracks):
        # Store the album names
        track_album.append(albums["title"][i])
        # Store the track names
        track_name.append(tracks_raw_infos[i]['items'][y]["name"])
        # Store the track ids
        track_id.append(tracks_raw_infos[i]['items'][y]["id"])
        # Store the track durations
        track_duration_ms.append(tracks_raw_infos[i]['items'][y]["duration_ms"])
        # Store the track types
        track_type.append(tracks_raw_infos[i]['items'][y]["type"])
        # Store the track urls
        track_url.append(tracks_raw_infos[i]['items'][y]["external_urls"]["spotify"])
        # Store the preview urls
        track_preview_url.append(tracks_raw_infos[i]['items'][y]["external_urls"]["spotify"])
        # Store the track expliciteness
        track_explicit.append(tracks_raw_infos[i]['items'][y]["explicit"])
        # Store the track number
        track_number.append(tracks_raw_infos[i]['items'][y]["track_number"])

    return {"name": track_name,
              "id": track_id,
              "album": track_album,
              "number": track_number,
              "duration_in_ms": track_duration_ms,
              "type": track_type,
              "url": track_url,
              "preview_url": track_preview_url,
              "explicit": track_explicit
      }

  def get_audio(self, tracks)->dict:

    audio_raw_infos = []

    # Display and store the audio infos
    audio_danceability = []
    audio_energy = []
    audio_key = []
    audio_loudness = []
    audio_mode = []
    audio_speechiness = []
    audio_instrumentalness = []
    audio_acousticness = []
    audio_liveness = []
    audio_valence = []
    audio_tempo = []
    audio_type = []
    audio_time_signature = []

    num_total_tracks = len(tracks['name'])
    # Store the audio infos of all the tracks
    for i in range(num_total_tracks):
      store_temp = self.sp.audio_features(tracks=tracks['id'][i])

      audio_raw_infos.append(store_temp)

    # I want to add the audio analysis
    for i in range(num_total_tracks):
      # Store the audio danceability
      audio_danceability.append(audio_raw_infos[i][0]["danceability"])
      # Store the audio energy
      audio_energy.append(audio_raw_infos[i][0]["energy"])
      # Store the audio key
      audio_key.append(audio_raw_infos[i][0]["key"])
      # Store the audio loudness
      audio_loudness.append(audio_raw_infos[i][0]["loudness"])
      # Store the audio mode
      audio_mode.append(audio_raw_infos[i][0]["mode"])
      # Store the audio speechiness
      audio_speechiness.append(audio_raw_infos[i][0]["speechiness"])
      # Store the audio instrumentalness
      audio_instrumentalness.append(audio_raw_infos[i][0]["instrumentalness"])
      # Store the audio acousticness
      audio_acousticness.append(audio_raw_infos[i][0]["acousticness"])
      # Store the audio liveness
      audio_liveness.append(audio_raw_infos[i][0]["liveness"])
      # Store the audio valence
      audio_valence.append(audio_raw_infos[i][0]["valence"])
      # Store the audio tempo
      audio_tempo.append(audio_raw_infos[i][0]["tempo"])
      # Store the audio type
      audio_type.append(audio_raw_infos[i][0]["type"])
      # Store the audio time signature
      audio_time_signature.append(audio_raw_infos[i][0]["time_signature"])

    return {"danceability": audio_danceability,
            "energy": audio_energy,
            "key": audio_key,
            "loudness": audio_loudness,
            "mode": audio_mode,
            "speechiness": audio_speechiness,
            "instrumentalness": audio_instrumentalness,
            "acousticness": audio_acousticness,
            "liveness": audio_liveness,
            "valence": audio_valence,
            "tempo": audio_tempo,
            "audio_type": audio_type,
            "time_signature": audio_time_signature
        }


  def create_dataset(self, dic)->pd.DataFrame:

    df = pd.DataFrame.from_dict(dic)

    return df
```

```python
# Infos about the parameters

# acousticness - A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic.

# danceability - Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable.

# energy - Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low on the scale. Perceptual features contributing to this attribute include dynamic range, perceived loudness, timbre, onset rate, and general entropy.

# instrumentalness - Predicts whether a track contains no vocals. "Ooh" and "aah" sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly "vocal". The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content. Values above 0.5 are intended to represent instrumental tracks, but confidence is higher as the value approaches 1.0.

# liveness - Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live.

# loudness - The overall loudness of a track in decibels (dB). Loudness values are averaged across the entire track and are useful for comparing relative loudness of tracks. Loudness is the quality of a sound that is the primary psychological correlate of physical strength (amplitude). Values typically range between -60 and 0 db.

# speechiness - detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks.

# tempo - The overall estimated tempo of a track in beats per minute (BPM). In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration.

# valence - A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).
```
